```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.normal(100, 5, 30)
```

```python
plt.plot(y, color = "red", ls = '--', marker = "o")
plt.title('Title')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot()
```

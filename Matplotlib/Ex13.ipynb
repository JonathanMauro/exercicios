```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)
data = np.random.randint(low = 100, high = [250, 300, 500, 1000], size = (12, 4))
mes =  pd.date_range(start = '2021-1-1', periods = 12, freq = 'M')
colunas = ['A', 'B', 'C', 'D']
```

```python
df = pd.DataFrame(data, index = mes.month_name(), columns = colunas)
df
```

```python
plt.figure(figsize= (15,5))
plt.plot(df)
plt.title('Values over time')
plt.legend(df.columns, loc = 'best')
plt.grid(linestyle = '--', alpha = 0.5)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.ylim(80,1000)
plt.savefig('fig13.jpeg')
plt.show()
```

Use os arrays a seguir para construir um gráfico de linhas. Dentro da função plt.plot() mude o marcador para 'o', (marker = 'o') e a cor da linha para 'black' (color = 'black').

https://prnt.sc/2d8RSgjhTZPE

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(20)
y = np.random.randint(1,100,20)
```

```python
plt.plot(x,y, marker = 'o', color = 'black' )

plt.title('Title')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.show()
```

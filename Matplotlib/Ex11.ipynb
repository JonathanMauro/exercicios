```python
import matplotlib.pyplot as plt
```

```python
sales = [1500.2, 1378, 671, 431, 700]
labels = ['Brand '+str(number) for number in range(1,len(sales)+1)]
```

```python
plt.pie(sales,labels = labels)
plt.savefig('figure11.jpeg')
plt.show()
```

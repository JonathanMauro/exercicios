Use os arrays a seguir para construir um gráfico de linhas. Use a biblioteca Matplotlib e configure o título com plt.title(), e a descrição dos eixos com plt.xlabel() e plt.ylabel().

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(-5.0, 5.0, 0.01)
y = np.sin(np.pi*x)

```

```python
plt.plot(x,y)
plt.title('Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.show()
```

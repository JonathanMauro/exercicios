```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A)
df = pd.read_csv('Iris.csv')
df
```

```python
data = df.drop(columns= 'Id', inplace= True)
```

```python
frequency_iris = df['Species'].value_counts()
```

```python
plt.bar(x = frequency_iris.index, height = frequency_iris)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylim(0,80)
plt.show()
```

```python
plt.barh(y = frequency_iris, width = frequency_iris)
plt.title('Frequency of the three species in the Iris Dataset')
plt.xlim(0, 80)
plt.show()
```

```python
#16)
explode = (0, 0.1, 0.1)
labels = frequency_iris.index
plt.pie(frequency_iris, explode = explode, shadow = True, autopct = '%1.2f%%', labels = labels)
plt.title('Pie chart', fontsize = 15)
plt.legend(labels, loc = 'upper left')
plt.show()
```

```python
#17)
plt.scatter(x = df['SepalWidthCm'], y = df['SepalLengthCm'])
plt.show()
```

```python
#18)
plt.scatter(x = df['PetalWidthCm'], y = df['PetalLengthCm'])
plt.show()
```

```python
#19)
for column in df.columns[:-1]:
    plt.hist(df[column], bins = 20)
    plt.title(column+' histogram')
    plt.show()
```

```python
#20)

for column in df.columns[:-1]:
    plt.boxplot(df[column])
    plt.title(column+' boxplot', fontsize = 12, pad = 10)
    plt.show()
```

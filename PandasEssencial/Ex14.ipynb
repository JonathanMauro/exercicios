Para o DataFrame anterior plot um gráfico de linhas e adicione uma nova coluna com o quadrado de cada número ímpar no intervalo [1,20]. Para tanto use: df.plot().

```python
import pandas as pd

dados = { 'A' : [i for i in range(1,21) if i%2 ==0],
         'B': [i**3 for i in range(1,21) if i%2 !=0]}

```

```python
dados['C'] = [i**2 for i in range(1,21) if i%2 !=0 ]
df = pd.DataFrame(dados, index = range(1, len(dados['A'])+1))
df.plot()
```

```python
import pandas as pd
import numpy as np
```

```python
dic = {
    'A': [11, 3, 7, 99, 100, 130, 150],
    'B': [1.2, 0.5, 1.2, 3.2, 0.9, 1.7, 9.5],
    'C': ['A', 'Z', 'W', 'K', 'A', 'Z', 'Z'],
    'D': [True, False, True, False, True, False, True],    
    'E': ['A', 'A', 'B', 'B', 'A', 'A', 'B'],
    'F': [True, True, False, False, True, False, True],}
```

```python
index = list(range(10,71,10))
df = pd.DataFrame(data = dic, index = index)


df
```

```python
#A)

print(f'Dtype de cada coluna é : \n {df.dtypes}')
```

```python
#B)

df[['C','E']] = df[['C', 'E']].astype('category')  #convertendo o dtype das colunas 'C' e 'E'
df.dtypes
```

```python
#C)
df.select_dtypes('category')
```

```python
#D)
df.select_dtypes('bool')
```

```python
#F)
df.select_dtypes(include = ['int64', 'float64'])
```

```python
#G)
df.describe(include = 'all').T
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
anos = np.arange(2020,1899,-10)
valores = [i**3 for i in range(1,14)]
valor = valores[::-1]

series = pd.Series(data = valor, index = anos, name = 'Crescimento no tempo')
series
```

```python
series.plot(title = 'Taxa de Crescimento no Tempo');
```

Gere uma distribuição normal padrão e utilize o módulo Seaborn para construir uma visualização dos dados. A amostra deve conter 3000 números. Use a função displot() do módulo Seaborn e importe também

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
```

```python
data = random.normal(size = 3000)
```

```python
sns.displot(data, kind='kde')
plt.show()
```

```python
sns.displot(data, kind ='hist')
plt.show()
```

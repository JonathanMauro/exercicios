Use o módulo Seaborn para visualizar a distribuição de números aleatórios oriunda de uma distribuição uniforme, de tamanho 1000.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dist = np.random.uniform(size = 1000)
sns.displot(dist, kind = 'kde')
plt.show()
```

```python
sns.displot(dist, kind = 'hist')
plt.show()
```

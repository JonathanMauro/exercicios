22

Faça um gráfico de linhas para demonstrar a evolução da variável 'Airline-passengers' no tempo.

```python
series.plot(color = 'black')
plt.title('International airline passengers')
plt.grid(alpha = 0.5)
plt.xlabel('Time')
plt.ylabel('monthly totals in thousands')
plt.show()
```

23

Faça um boxplot caracterizar a variável 'Airline-passengers'.

```python
sns.boxplot(data =series)
plt.title('Boxplot')
plt.xlabel('Airline passengers')
plt.show()
```

24
Faça um histograma e um kde plot.

```python
sns.histplot(series)
plt.title('Airline passengers - Histogram')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(series)
plt.title('Kernel density estimation')
plt.show()
```

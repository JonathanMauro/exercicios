Carregue o dataset Iris anexo nesta aula;

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df
```

Remova a coluna Id;

```python
df.drop(columns= 'Id', inplace= True)
df
```

Use a função countplot() para mostrar a frequência das três espécies de Iris do dataset.

```python
sns.countplot(x = 'Species', data = df)
plt.title('Frequency of the three species in the Iris Dataset')
plt.ylabel('Frequency')
plt.ylim(0,80)
```

Faça um scatter plot para mostrar a relação entre 'SepalLengthCm' e 'SepalWidthCm';

```python
sns.scatterplot(x = df['SepalWidthCm'], y = df['SepalLengthCm'], data = df, hue = 'Species', alpha = 0.5)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize = 12)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

Faça um scatter plot para mostrar a relação entre 'PetalLengthCm' e 'PetalWidthCm'.

```python
sns.scatterplot(x = df['PetalWidthCm'], y = df['PetalLengthCm'], data = df, hue = 'Species', alpha  = 0.5)
plt.title('Relationship between Petal Length and Petal Width', fontsize = 12)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

3
Faça um histograma das features do dataset (exclusive coluna Species).

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, figsize = (10, 10))

sns.histplot(x = 'SepalLengthCm', data = df, ax = ax1)
ax1.set(title = 'Sepal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'SepalWidthCm', data = df, ax = ax2)
ax2.set(title = 'Sepal Width histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalLengthCm', data = df, ax = ax3)
ax3.set(title = 'Petal Length histogram', ylabel = 'Frequency')

sns.histplot(x = 'PetalWidthCm', data = df, ax = ax4)
ax4.set(title = 'Petal Width histogram', ylabel = 'Frequency')

plt.tight_layout(pad = 2)
```

Faça um boxplot das features do dataset (exclusive coluna Species).

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df)
plt.title('Box plot')
plt.show()
```

Faça um heatmap das features do dataset (exclusive coluna Species);

```python
corr = df.corr()  #Compute pairwise correlation of columns
```

```python
sns.heatmap(corr, cmap = 'Blues', annot = True)
```

Faça um pairplot para analisar a correlação entre as variáveis do dataset.

```python
sns.pairplot(data = df)
plt.show()
```

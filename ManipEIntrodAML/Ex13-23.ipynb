15. Gere um gráfico para a proporção entre os gêneros;

```python
df['Gender'].value_counts().plot(kind = 'bar', figsize = (8, 5))
plt.title('Proporção entre os gêneros', fontsize= 15)
plt.xlabel('Gender')
plt.xticks([0, 1], ['Feminino', 'Masculino'], rotation = 0)
plt.ylabel('Quantidade')
plt.xlabel('Gênero')
plt.show()
```

17. Gere um gráfico de barras para mostrar a distribuição do nível de escolaridade dos pais por gênero;

```python
plt.figure(figsize = (10, 5))
sns.countplot(
    data = df,
    x = 'Parental_Level_Of_Education',
    hue= 'Gender'
)
plt.title('Nível de escolaridade parental por gênero', pad = 20)
plt.tight_layout();
```

20. Gere um gráfico com subplots com 3 colunas, em que cada plot demonstra a distribuição da nota das provas de acordo com a realização ou não do curso preparatório;

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols = 3, figsize = (15,6))

sns.kdeplot(data = df, x = 'Math_Score', hue = 'Test_Preparation_Course', ax = ax0)
ax0.set_xlabel('Math Score')

sns.kdeplot(data = df, x = 'Reading_Score', hue = 'Test_Preparation_Course', ax = ax1)
ax1.set_xlabel('Reading Score')

sns.kdeplot(data = df, x = 'Writing_Score', hue = 'Test_Preparation_Course', ax = ax2)
ax2.set_xlabel('Writing Score')

plt.tight_layout(pad = 3.5)
```

22. Gere um mapa de calor para analisar a relação entre as features do dataset;

```python
mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(
    correlation_matrix,
    annot = True,
    mask = mask,
    square = True,
    cmap = 'Blues');
```
